### Feature Engineering

```python
dfc = data_awards[(data_awards["outcome"] == "Nominated") | (data_awards["outcome"] == "Won")].reset_index(drop=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
enc_data=pd.DataFrame(ohe.fit_transform(dfc[["outcome"]]).toarray(), columns=["Nominated", "Won"])
```

```python
dfc = dfc.join(enc_data)
```

```python
data_awards_cleaned = dfc.groupby(["director_name", "year"], as_index=False).aggregate({"Nominated": "sum", "Won": "sum"})
data_awards_cleaned
```

```python
data_awards_cleaned[["Nominated_CumSum","Won_CumSum"]] = data_awards_cleaned.groupby("director_name")["Nominated", "Won"].transform(pd.Series.cumsum)
```
